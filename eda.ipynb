```python
import pandas as pd
import numpy as np
import spacy
from collections import Counter
import json

import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
def import_book(f_name):
    with open(f_name, 'r', encoding='utf-8-sig') as f:
        book = f.read()
    
    book =  book.replace('\n', ' ')
    pages = book.split("________________")
    return pages, ' '.join(pages)
```

```python
def get_people(item):
    if item[1] == 'PERSON':
        return True
    else:
        return False
```

```python
def get_occuring_sent(word):
    sentences = [sentence + '.' for sentence in book.split('.') if word in sentence]
    return sentences
```

```python
pages, book = import_book("data/cains_jawbone.txt")
```

```python
doc = nlp(book)
```

```python
entities = [(X.text, X.label_) for X in doc.ents]
```

```python
people = list(filter(get_people, entities))
people = [person[0] for person in people]
```

```python
unique_people = np.unique(people)
```

```python
people_context = {person: get_occuring_sent(person) for person in unique_people}
```

```python
sorted_people = dict(sorted(people_context.items(), key=lambda x: len(x[1]), reverse=True))
```

```python
for k,v in sorted_people.items():
    print("Person:", k)
    print("------\n")
    for sent in v:
        print(sent)
    print("----------\n")
```

```python
with open("data/people_context.json", "w", encoding='utf-8-sig') as outfile:
    json.dump(sorted_people, outfile)
```
